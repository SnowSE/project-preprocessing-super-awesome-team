## Part 3: Data Preprocessing

# Variable Removal
After looking at the EDA for the data, there are candidates for removal, but we have not removed them at this time because there is some variance that may or may not contain a signal when we begin modeling. At a later time we may revisit our decision.  But looking at our correlation matrix, none of the variables has a very strong correlation with loneliness. Some candidates for removal:

| Variable Name | Reason for possible removal |
|--------------|-----------|
| Music | Nearly all responses are identical offering little predictive value |
| Movies | Nearly all responses are identical offering little predictive value |
| Siblings |Nearly all responses are identical offering little predictive value | 

For now, however, we will maintain all 15 variables for modeling. 

# Handling Missing Values
As illustrated earlier (box 3) the largest missing value column was 7 / 1010 which is less than 1%.  As such no columns will be eliminated for missing values. We then determined if a row was missing more than 2 items it was more than 10% incomplete and we would remove those rows. Examining each row, there was only 1 row that met this criteria, and it was removed (see box 7). There was also one row in which the output variable was missing and we did not want to impute the output variable, so this row was removed.  We then used different imputers to fill in the remaining missing values based on the following chart:

| Variable Name | Variable Type | % Missing | Imputation Method | Justification |
|--------------|-----------|-------------|----------|---------------|
| Music | Numerical | 0.3% | Median | Data showed right skew in EDA|
| Techno | Numerical | 0.4% | Median | Data showed left skew in EDA|
| Movies | Numerical | 0.6% | Median | Data showed right skew in EDA|
| History | Numerical | 0.2% | Median | Data showed left skew in EDA|
| Mathematics | Numerical | 0.3% | Median | Data showed left skew in EDA|
| Pets | Numerical | 0.4% | Median | Median most forgiving imputer|
| Spiders | Numerical | 0.3% | Median | Median most forgiving imputer|
| Parents Advice | Numerical | 0.2% | Median | Median most forgiving imputer|
| Finances | Numerical | 0.3% | Median | Median most forgiving imputer
| Age | Numerical | 0.7% | Median | Median most forgiving imputer|
| Siblings | Numerical | 0.6% | Median | Median most forgiving imputer|
| Gender | Categorical | 0.6% | Mode | Imputing Categorical Data|
| Village_town | Categorical | 0.4% | Mode | Imputing Categorical Data|

# Encoding Categorical Variables
| Variable Name | Variable Type | Encoding Method | Justification | Resulting Columns |
|--------------|-----------|-------------|----------|---------------|
| Internet Usage | Ordinal | Ordinal-Encoding | There was a clear order to the different options | 1 Column |
| Gender | Nominal | One-Hot | There is no inherent order | 1-Column (Gender_Male) |
| Village_town | Nominal | One-Hot | There is no inherent order | 1-Column (Village-town_village) |

```python
## Ruberic says ... Remove columnts > 10 % missing.  The largest deficit is 7 / 1010, nowhere near 10% all columns stay

## Before we examine the rows, the one row where 'loneliness' is missing will just be eliminated. We cannot predict that which is missing 

df = df.dropna(subset=['Loneliness'])

## Remove rows with > 10% missing Checking for rows.  ! missing item is < 10%, 2 missing items is > 10%

missing_count_per_row = df.isnull().sum(axis=1) 
print(f"Rows with > 10% missing: {(missing_count_per_row > 2).sum()}")

problem_children = df[missing_count_per_row > 2]
print(problem_children)

df = df.drop(problem_children.index)

# Impute the remaining missing values: (First the numerical values)

num_imputer = SimpleImputer(strategy='median') # median is more resiliant, especialy with data that is slightly skewed
df[numerical_cols] = num_imputer.fit_transform(df[numerical_cols])

cat_imputer = SimpleImputer(strategy='most_frequent') # mode for categorical data
df[categorical_cols] = cat_imputer.fit_transform(df[categorical_cols])
```

```python
# encoding categorical values. 

# First only 'gender' and 'Village - town' are actually categorical
df = pd.get_dummies(df, columns = ['Gender', 'Village - town'], drop_first=True, dtype=int)

# Internet Usage is ordinal, just without the numbers. 
internet_mapping = {
    'no time at all': 0,
    'less than an hour a day': 1,
    'few hours a day': 2,
    'most of the day': 3
}
df['Internet usage'] = df['Internet usage'].map(internet_mapping).astype(int)
```

## Part 4: Cross Validation (Train-Test Split)

# Split Strategy
Our split strategy is supported by many other researchers. The debate is between an 80/20 split or a 70/30 split. The recommendation from management requested we consider a 70/30 split. We introduced a random state of 42 for reproducibility and consistency.  This affords sufficient data to adequately train the data with a healthy reserve of data to test the models ability to predict. 

# Dataset Dimensions Table
| Dataset | Samples | Features | % of Total |
|---------|---------|----------|------------|
| Training | 705 | 15 | 70% |
| Testing | 303 | 15 | 30% |

While the variables chosen at randome will be sufficient, we stratified against the output variable (loneliness) to ensure equal numbers of 1, 2, 3, 4, and 5 in the two sets of data.  The success of this stratification can be seen in the graph included with box 15. There is clearly equal numbers of each response represented in our split data. 

```python
# Cross validation , 70/30 per instructions

X = df.drop('Loneliness', axis=1)
y = df['Loneliness']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,   # for 70/30 split
    random_state=42,  # for random selection 
    stratify=y       # for equal distribution of 'y' values in test and train
)

print(f"Training: {X_train.shape[0]} samples, {X_train.shape[1]} features")
print(f"Testing: {X_test.shape[0]} samples, {X_test.shape[1]} features")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,5))

train_counts = y_train.value_counts().sort_index()
ax1.bar(train_counts.index, train_counts.values, color='steelblue', edgecolor='black')
ax1.set_xlabel('Loneliness level')
ax1.set_ylabel('Frequency')
ax1.set_title('Training Set Target Distribution')
ax1.set_xticks([1,2,3,4,5])

test_counts = y_test.value_counts().sort_index()
ax2.bar(test_counts.index, test_counts.values, color='coral', edgecolor='black')
ax2.set_xlabel('Loneliness level')
ax2.set_ylabel('Frequency')
ax2.set_title('Test Set Target Distribution')
ax2.set_xticks([1,2,3,4,5])
fig.text(0.5, 0.005, 'Figure Caption: Target distributions remain balanced across training and test sets', 
         ha='center', fontsize=11)

plt.tight_layout()
plt.show()
```

## Part 5: Feature Scaling

# Scaling Decision Table
Before proceeding with which variables to scale, we decided that standardization was appropriate for our data as it is the most widely accepted method for pre-processing on most modeling systems, and was best able to scale for skewed data or data with outliers. As such we created a standard scaler and applied it according to the following table: 

| Feature Name | Data Type | Scaling Applied? | Justification |
|--------------|-----------|------------------|---------------|
| Music | Ordinal (1-5) | Yes | Range 1-5, need standardization |
| Techno | Ordinal (1-5) | Yes | Range 1-5, need standardization |
| Movies | Ordinal (1-5) | Yes | Range 1-5, need standardization |
| History | Ordinal (1-5) | Yes | Range 1-5, need standardization |
| Mathematics | Ordinal (1-5) | Yes | Range 1-5, need standardization |
| Pets | Ordinal (1-5) | Yes | Range 1-5, need standardization |
| Spiders | Ordinal (1-5) | Yes | Range 1-5, need standardization |
| Parents' advice | Ordinal (1-5) | Yes | Range 1-5, need standardization |
| Internet usage | Ordinal (0-3) | Yes | Range 0-3, need standardization |
| Finances | Ordinal (1-5) | Yes | Range 1-5, need standardization |
| Age | Continuous | Yes | Range 17-24, need standardization |
| Siblings | Discrete | Yes | Range 0-10+, need standardization |
| Gender_male | Binary | No | Already 0/1 |
| Village-town_village | Binary | No | Already 0/1 |

We then applied mean and standard deviation to the final data set and as expected the average was 0 and the standard deviation of the final variables was 1 (see table 21). 

```python
# we will scale the non-binary variables, and not touch the binary variables

binary_cols = ['Gender_male','Village - town_village']
non_binary_cols = [col for col in X_train.columns if col not in binary_cols]

scaler = StandardScaler()
scaler.fit(X_train[non_binary_cols])

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[non_binary_cols] = scaler.transform(X_train[non_binary_cols])
X_test_scaled[non_binary_cols] = scaler.transform(X_test[non_binary_cols])

print(X_train_scaled[non_binary_cols].describe().loc[['mean', 'std']].round(3))
```
